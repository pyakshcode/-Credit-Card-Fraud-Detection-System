```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a pandas Datafrem
credit_card_data = pd.read_csv('/content/sample_data/creditcard.csv')
```

```python
# first five row of the data set
credit_card_data.head()
```

```python
#dataset information
credit_card_data.info()
```

```python
# check the number of missing value in each coloum
credit_card_data.isnull().sum()
```

```python
#distribution of legit transactions & fraudulent transation
credit_card_data['Class'].value_counts()
```

this data set is highly unblanced

0 --- Normal transaction
1 --- Fraudulent transaction

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#staistiv=cal measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the value for both transaction
credit_card_data.groupby('Class').mean()
```

build  a sample dataset containing similat distribution of normal transactions and fraudulent transaction

Number of fraudulent transactions -- 165

```python
legit_sample = legit.sample(n=165)
```

Concatenating two Datafreams

```python
new_dataset = pd.concat([legit_sample, fraud ], axis =0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting data into features & targets

```python
X = new_dataset.drop(columns = 'Class' , axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

split the data into training data & testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

model training


Logistic Regression()

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train, Y_train)
```

model evaluation

accuracy score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data :- ' , training_data_accuracy)
```

```python
#accuracy on the data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction , Y_test)
```

```python
print('Accuract score on test data ', test_data_accuracy)
```
